```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = "https://fataniestate.com/"
```

```python
res = requests.get(url).text
```

```python
res
```

```python
soup = BeautifulSoup(res, 'html.parser')
```

```python
print(soup.prettify())
```

```python
bad = []
good = []
keywords = []
```

```python
#title
title = soup.find('title').text
if title:
    good.append(f"Title Exists: {title}")
else:
    bad.append("No Title!")
    
    
#Meta
meta_d = soup.find('meta', attrs={'name': 'description'})['content']
if meta_d:
    good.append(f"Title Exists: {meta_d}")
else:
    bad.append("No Title!")
    
#headings
# Grab the Headings
hs = ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']
h_tags = []
for h in soup.find_all(hs):
    good.append(f"{h.name}-->{h.text.strip()}")
    h_tags.append(h.name)

if 'h1' not in h_tags:
    bad.append("No H1 found!")


# hs = ['h1', 'h2', 'h3']
# h_tags = []
# for h in soup.find_all(hs):
#     good.append(f"{h.name}-->{h.text.script()}")
#     h_tags.append(h.name)

# if 'h1' not in h_tags:
#     bad.append("No H1 Found!")
    
#Image ALt
for i in soup.find_all('img', alt=''):
    bad.append(f"No Alt: {i}")

print(good)
print('---------')
print(bad)
```
